## Snapshot


## 1. How does subscribers and customers behave differently?

   Subscribers take rides in both morning and commute time; whereas customers' rides take place most during weekend. And that represents the main purpose in different user type. 
   
   
   
## 2. Which age group favors e-bike more?

   Surrpinsgly, age 10-20 years age group seems to show the most interest in the e-bike. So younger a rider is, the more likely    the rider would be fond of electric bike rides.
   
   
  
## 3. What are the most popular docks for 'bike share for all' members?

   It is a subscription model for low income bay area residents. Unlike most popular docks. The popular docks for 'bike share      for all' members are spreaded especially around bart station, even though that is quite further far from market street.        Guess, it's because low income people tend to travel further from south bay to east bay. And they might take public  transportation.
   




## 4. Top 100 most frequently used paths

   The 100 paths of all bikes rides show that the most rides happened around SOMA(South of Market). In addition, there are lot      of rides around bart station. The top paths show that people take bike rides from station, such as Bart to where most of the    companies are located. And seems people take bike to go to restaurant around lunch time as well.
